```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import folium as fm
sns.set() 
```

# Load The Data

```python
df = pd.read_csv(r'C:\Users\nadav\OneDrive\שולחן העבודה\Data Analyst\Python Files\Dubai Apartments\properties_data.csv')
pd.set_option('display.max_columns',None)
df.head()
```

# Add New Column 'total_rooms'

```python
df['total_rooms'] = df['no_of_bedrooms'] +df['no_of_bathrooms']
df.head()
```

# Add New Column 'Price per Room'

```python
df['Price Per Room']=df['price']/df['total_rooms']
df.head()
```

# Making A Groupby for Neighborhoods

```python
df_grouped_by= df.groupby(['neighborhood']).mean()
df_grouped_by.head()
```

#### Change the neighborhood column index.

```python
df_grouped_by_2= df.groupby(['neighborhood'],as_index = False).mean().sort_values('price',ascending=False)
df_grouped_by_2.head()
```

```python
df_grouped_by.shape
```

#### 54 different neighborhoods in the data .

```python
df1=df_grouped_by.sort_values('price',ascending=False)
df1.head()
```

#### The most expensive neighborhood in Dubai is Palm Jumeirah	with  mean price for apartment of 4,379,435 \$ .

```python
df2=df_grouped_by.sort_values('price',ascending=True)
df2.head()
```

# Bar Plot of the Findings

```python
ax = df1['price'].plot.bar(color='#008B8B',figsize=(20,10))
ax= sns.set_style('whitegrid')
ax= plt.title('Mean Price for Apartment per Neighborhood in Millions',size= 30, fontweight='bold')
ax = plt.xlabel('neighborhood', size = 20,fontweight='bold')
ax=plt.ylabel('price',size = 20,fontweight='bold')
ax =sns.despine(left=True)
ax = plt.xticks(rotation=90, size = 16)
ax= plt.yticks(size = 16)
```

# More Digging

```python
plt.figure(figsize=(10,8))
ax= sns.set_style('whitegrid')
ax= plt.title('Count Of Apartments per Total rooms ',size= 20, fontweight='bold')
ax=sns.histplot(data=df, x="total_rooms",color='#008B8B',binwidth=0.4,kde=True)
ax.lines[0].set_color('black')
ax = plt.xticks(size=14)
ax = plt.yticks(size = 14)
ax = plt.xlabel('Total Rooms',size= 16)
ax = plt.ylabel('Count',size= 16)
```

#### We know from before that the average number of total rooms is 4

```python
plt.figure(figsize=(10,8))
ax= sns.set_style('whitegrid')
ax= plt.title('Count Of Apartments per quality ',size= 20, fontweight='bold')
ax=sns.histplot(data=df, x="quality",color='#008B8B',binwidth=0.4,kde=True)
ax.lines[0].set_color('black')
ax = plt.xticks(size=14)
ax = plt.yticks(size = 14)
ax = plt.xlabel('Quality of Apartment',size= 16)
ax = plt.ylabel('Count',size= 16)
```

#### Twice as much medium quality apartments than low quality apartments.

```python
plt.figure(figsize=(10,8))
ax= sns.set_style('whitegrid')

plt.xticks(size = 14)
plt.yticks(size = 14)
plt.title("Correlation Between Total Rooms and Price",
          size=20,
          fontweight='bold')
plt.xlabel('total_rooms',size=16)
ax= plt.ylabel(' Price in Millions',size= 16)
ax=sns.regplot(x='total_rooms',y ='price',data=df,color='#008B8B',ci=None)
plt.xlim(0,12)
```

```python
np.corrcoef(df['total_rooms'],df['price'])
```

#### The Correlation between the price and the total rooms is preety week(0.53). That's not common in the real estate industry

```python

plt.figure(figsize=(10,8))
ax= sns.set_style('whitegrid')
ax=sns.regplot(x='size_in_sqft',y ='price',data=df,color='#008B8B',ci=None)
plt.xticks(size = 14)
plt.yticks(size = 14)
#ax.set(xlabel='size_in_sqtf',ylabel='Price')
plt.title("Correlation Between size in sqtf and Price",
          size=20,
          fontweight='bold')
ax=plt.ylabel(' Price in Millions',size= 16)
ax = plt.xlabel('size_in_sqtf',size=16)
plt.xlim(100,10000)
```

```python
np.corrcoef(df['size_in_sqft'],df['price'])
```

#### As I was expecting, there is a strong postive correlation(0.81) between the size of the apartment and the price

# Let's Create an Heatmap to see wide information about the correlations 

```python
plt.subplots(figsize=(15, 12))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot = False,cmap="YlGnBu")
plt.show()
```

### From the heatmap we can conclude that the main variable that has  high correlation with the price  is the size of the apartment. The size in sqtf is the main factor to the apartment price.

```python
plt.figure(figsize=(9,8))
ax= sns.set_style('whitegrid')
ax=sns.boxplot(x=df["quality"], y=df["total_rooms"],color='#008B8B',meanline=True,notch = True)
ax=plt.title("Number of total rooms per apartment quality",
          size=20,
          fontweight='bold')
#plt.style.use('fivethirtyeight')
ax=plt.xticks(size = 14)
ax=plt.yticks(size = 14)
ax = plt.xlabel('Quality',size=16)
ax = plt.ylabel('Total Rooms',size=16)
```

#### Here is some intersting inforamtion from the boxplot:
- 50% of the Medium, Low and Ultra apartments have 4 rooms in total.
- 25% of the High quality apartments have more than 6 rooms.
- The range of total rooms in high quality apartments is wide, 1-10 rooms.
